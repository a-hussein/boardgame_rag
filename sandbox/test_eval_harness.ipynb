```python
import pickle, pandas as pd
from pathlib import Path
```

```python
with open("../indices/bm25.pkl", "rb") as f:
    obj = pickle.load(f) # dictionary with bm25 object, tokenized words, doc_id
```

```python
bm25 = obj["bm25"]
tokenized = obj["tokenized"]
doc_ids = obj["doc_ids"]
```

```python
print(f"Docs indexed: {len(doc_ids)}")
print(f"Example tokens for first 3 docs:")
for i in range(3):
    print(f"{doc_ids[i]}: {tokenized[i][:20]}")
```

```python
print("\nSample term frequencies for a few tokens:")
for term in list(bm25.idf.keys())[:10]:
    print(f"'{term}' → appears in {bm25.doc_freqs[term]} docs (IDF={bm25.idf[term]:.2f})")
```
